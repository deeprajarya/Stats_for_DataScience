```python
import numpy as np

mu =0.5
sigma =0.2
normal_distributed_data = np.random.normal(mu,sigma,1000)
normal_distributed_data
```

```python
sns.histplot(normal_distributed_data, kde=True)
```

```python
# generate log normal data

log_normal_distributed_data = np.random.lognormal(mu,sigma, 1000)
log_normal_distributed_data
```

```python
sns.histplot(log_normal_distributed_data,kde=True)
```

```python
mu=3.0
sigma=1.0
log_normal_distributed_data = np.random.lognormal(mu,sigma, 1000)
sns.histplot(log_normal_distributed_data,kde=True)


# This distribution is log normal distribution. Log normal and right skewed both are same.
```

```python

sns.histplot(np.log(log_normal_distributed_data),kde=True)
```

```python
import matplotlib.pyplot as plt
import scipy.stats as stat
import pylab


def plot_data(data):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    sns.histplot(data,kde=True)
    plt.subplot(1,2,2)
    stat.probplot(data,dist='norm',plot=pylab)
    plt.show()
```

```python
plot_data(log_normal_distributed_data)
```

```python
plot_data(np.log(log_normal_distributed_data))
```
